Create a database


Create a database

```python
import sqlite3

conn = sqlite3.connect('workforce.db')
cursor = conn.cursor()

print("Database created successfully")
```

Create the table

```python
cursor.execute("""
CREATE TABLE employee_workforce (
EmployeeID INTEGER PRIMARY KEY,
EmployeeName TEXT,
Department TEXT,
Age INTEGER,
Gender TEXT,
Salary INTEGER,
PerformanceScore REAL,
Experience INTEGER,
Attrition TEXT,
TrainingHours INTEGER
)
""")

conn.commit()

print("Table created successfully")
```

Insert the data

```python
cursor.executemany("""
INSERT INTO employee_workforce
VALUES (?,?,?,?,?,?,?,?,?,?)
""",[
(101,'Rahul','HR',28,'Male',35000,4.5,3,'No',20),
(102,'Priya','IT',25,'Female',45000,4.8,2,'No',18),
(103,'Arun','Finance',30,'Male',50000,4.2,5,'Yes',15),
(104,'Sneha','Marketing',27,'Female',40000,4.6,4,'No',22),
(105,'Kiran','IT',29,'Male',55000,4.9,6,'No',25)
])

conn.commit()

print("Data inserted successfully")
```

Display records

```python
cursor.execute("SELECT * FROM employee_workforce")

rows = cursor.fetchall()

for row in rows:
    print(row)
```

Perform the ALTER TABLE operation to add a new column named Email to the employee_workforce table.

```python
cursor.execute("""
ALTER TABLE employee_workforce
ADD COLUMN Email TEXT
""")

conn.commit()

print("Column added successfully")
```

Update the salary of the employee whose EmployeeID is 101.

```python
cursor.execute("""
UPDATE employee_workforce
SET Salary = 60000
WHERE EmployeeID = 101
""")

conn.commit()

print("Salary Updated Successfully")
```

```python
cursor.execute("SELECT * FROM employee_workforce WHERE EmployeeID=101")

print(cursor.fetchall())
```

Display employee records in ascending order of salary.

```python
cursor.execute("""
SELECT *
FROM employee_workforce
ORDER BY Salary ASC
""")

rows=cursor.fetchall()

for row in rows:
    print(row)
```

Group employees by department and count the total number of employees in each department.

```python
cursor.execute("""
SELECT Department,
COUNT(*) AS TotalEmployees
FROM employee_workforce
GROUP BY Department
""")

rows=cursor.fetchall()

for row in rows:
    print(row)
```

Find the highest salary among all employees.

```python
cursor.execute("""
SELECT MAX(Salary)
FROM employee_workforce
""")

print(cursor.fetchall())
```

Delete the employee whose EmployeeID is 110.

```python
cursor.execute("""
DELETE FROM employee_workforce
WHERE EmployeeID=110
""")

conn.commit()

print("Record Deleted Successfully")
```

Create a Departments table and insert department manager details.

```python
cursor.execute("""
CREATE TABLE departments(
DepartmentID INTEGER PRIMARY KEY,
Department TEXT,
Manager TEXT
)
""")

cursor.executemany("""
INSERT INTO departments
VALUES(?,?,?)
""",[
(1,'HR','Anita'),
(2,'IT','Rahul'),
(3,'Finance','Kumar'),
(4,'Marketing','Priya'),
(5,'Sales','Arun')
])

conn.commit()

print("Departments Table Created")
```

Add DepartmentID Column

```python
cursor.execute("""
ALTER TABLE employee_workforce
ADD COLUMN DepartmentID INTEGER
""")

conn.commit()

print("DepartmentID column added.")
```

Update the DepartmentID values

```python
cursor.execute("UPDATE employee_workforce SET DepartmentID=1 WHERE EmployeeID=101")
cursor.execute("UPDATE employee_workforce SET DepartmentID=2 WHERE EmployeeID=102")
cursor.execute("UPDATE employee_workforce SET DepartmentID=3 WHERE EmployeeID=103")
cursor.execute("UPDATE employee_workforce SET DepartmentID=4 WHERE EmployeeID=104")
cursor.execute("UPDATE employee_workforce SET DepartmentID=2 WHERE EmployeeID=105")

conn.commit()
```

```python
cursor.execute("SELECT * FROM departments")
print(cursor.fetchall())
```

Perform an INNER JOIN between the employee_workforce and departments tables using DepartmentID.

```python
cursor.execute("""
SELECT e.EmployeeID,
       e.EmployeeName,
       e.Department,
       d.Manager
FROM employee_workforce e
INNER JOIN departments d
ON e.DepartmentID=d.DepartmentID
""")

rows=cursor.fetchall()

for row in rows:
    print(row)
```

Drop the employee_workforce table from the database.

```python
cursor.execute("DROP TABLE employee_workforce")
conn.commit()

print("Table dropped successfully.")
```

Close the database connection

```python
conn.close()

print("Database connection closed successfully.")
```
