This Python project demonstrates fundamental programming concepts using an Employee Workforce dataset. It covers Object-Oriented Programming (OOP), Lists, and Dictionaries to organize, manage, and access employee information efficiently.

Create a simple Employee class using OOP and display employee details.

```python
class Employee:
    def __init__(self, name, department):
        self.name = name
        self.department = department

emp = Employee("Rahul", "HR")

print("Employee Name:", emp.name)
print("Department:", emp.department)
```
